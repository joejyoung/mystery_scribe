```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
scoreboard = cv2.imread("../data/raw/sample.png")
```

```python
scoreboard = cv2.cvtColor(scoreboard, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(scoreboard)
```

```python
scoreboard_gray = cv2.cvtColor(scoreboard, cv2.COLOR_BGR2GRAY)
plt.imshow(scoreboard_gray, cmap="gray")
```

```python
gray = np.float32(scoreboard_gray)
```

Harris Corner Detection

```python
dst = cv2.cornerHarris(src=gray, blockSize=2, ksize=3, k=0.04) # Harris Corner Detection
```

```python
dst = cv2.dilate(dst, None); dst.max()
```

```python
scoreboard[dst > 0.01 * dst.max()] = [255, 0, 0]
```

```python
plt.imshow(scoreboard)
```

Shi-Tomasi Corner Detector

```python
corners = cv2.goodFeaturesToTrack(scoreboard_gray, 5, 0.01, 10)
corners = np.int8(corners)
```

```python
scoreboard_copy = scoreboard.copy()
for i in corners:
    x, y = i.ravel()
    cv2.circle(scoreboard_copy, (x, y), 3, 255, -1)
plt.imshow(scoreboard_copy)
```

```python
corners = cv2.goodFeaturesToTrack(scoreboard_gray, 64, 0.01, 10)
corners = np.int8(corners)
```

```python
scoreboard_copy = scoreboard.copy()
for i in corners:
    x, y = i.ravel()
    cv2.circle(scoreboard_copy, (x, y), 3, 255, -1)
```

```python
cv2.imwrite("../data/processed/roi_corner.png", cv2.cvtColor(scoreboard_copy, cv2.COLOR_BGR2RGB))
plt.imshow(scoreboard_copy)
```
